```python
import pandas as pd
```

```python
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```python
cnts.head(10)
```

```python
# Drop rows with NaN values in 'clip_enrichment' and extract the 'geneid' values
clip_enrichment_geneids = cnts.dropna(subset=['clip_enrichment']).index.tolist()

# Drop rows with NaN values in 'rden_change' and extract the 'geneid' values
rden_change_geneids = cnts.dropna(subset=['rden_change']).index.tolist()
```

```python
print("First 5 Gene IDs with non-NaN 'clip_enrichment':", clip_enrichment_geneids[:5])
```

```python
# Convert lists to pandas Series to use string operations
clip_enrichment_series = pd.Series(clip_enrichment_geneids)
rden_change_series = pd.Series(rden_change_geneids)

# Perform the string split and extraction
clip_enrichment_id2 = clip_enrichment_series.str.split('.').str.get(0).tolist()
rden_change_id2 = rden_change_series.str.split('.').str.get(0).tolist()
```
